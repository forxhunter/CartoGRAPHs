# DISEASE association Matrix

```python
all_doids = list(d_do_genes.keys())

subdata = d_gene_do # dict(it.islice(d_gene_do.items(),G.number_of_nodes()))
genes = list(subdata.keys())
genes_doids = subdata.values()

# Get a list of all matches = search for matches of Disease ID list and gene list 
matches = []
for i in genes_doids:
    gene_i = []
    for j in i:
        for idx,value in enumerate(all_doids):
            if j == value:
                gene_i.append(idx)
    matches.append(gene_i)
    
all_match = []
for sub in matches:
    gene_match = [0] * len(all_doids)
    for elem in sub:
        for idx,value in enumerate(gene_match):
            if elem == idx:
                gene_match[idx] = 1
    all_match.append(gene_match)   
    
all_matrix = np.row_stack(all_match)
```

```python
df_DIS = pd.DataFrame(all_matrix, columns = all_doids, index = genes, dtype = int)
```

```python
%%time 

metric = "cosine" 
DM_disease = pd.DataFrame(distance.squareform(distance.pdist(df_DIS, metric)), index=genes, columns=genes)
```
